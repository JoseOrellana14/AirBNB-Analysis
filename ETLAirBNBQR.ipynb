```python
# pip install pandas

import pandas as pd

file_path = "BBDD Huespedes (Macros).xlsm"

# Leer las dos hojas específicas en un diccionario de DataFrames
sheets_dict = pd.read_excel(file_path, sheet_name=["BBDD CoHost", "BBDD Anfitrion"], engine="openpyxl")

# Extraer cada hoja en un DataFrame separado
df_cohost = sheets_dict["BBDD CoHost"]
df_anfitrion = sheets_dict["BBDD Anfitrion"]

# Mostrar las primeras filas de cada DataFrame para verificar
print("BBDD CoHost:")
print(df_cohost.head())
df_cohost.info()

print("\nBBDD Anfitrion:")
print(df_anfitrion.head())

```

```python
#################    CoHost    ###############

# Convirtiendo la columna Ingreso de object a date Tanto para df_cohost
# Create a copy of cleaned_df_cohost to avoid modifying the original dataframe
cleaned_df_cohost = df_cohost.copy()

# Convertir la columna 'Ingreso' a formato datetime
cleaned_df_cohost['Ingreso'] = pd.to_datetime(cleaned_df_cohost['Ingreso'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_cohost['Ingreso'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Ingreso'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_cohost['Ingreso'].dtype)




#################    Anfitrion    ###############
# Convirtiendo la columna Ingreso de object a date Tanto para df_anfitrion
# Create a copy of cleaned_df_anfitrion to avoid modifying the original dataframe
cleaned_df_anfitrion = df_anfitrion.copy()

# Convertir la columna 'Ingreso' a formato datetime
cleaned_df_anfitrion['Ingreso'] = pd.to_datetime(cleaned_df_anfitrion['Ingreso'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_anfitrion['Ingreso'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Ingreso'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_anfitrion['Ingreso'].dtype)
```

```python
#################    CoHost    ###############
# Convirtiendo la columna Salida de object a date en cleaned_df_cohost

# Convertir la columna 'Salida' a formato datetime
cleaned_df_cohost['Salida'] = pd.to_datetime(cleaned_df_cohost['Salida'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_cohost['Salida'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Salida'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_cohost['Salida'].dtype)



#################    Anfitrion    ###############
# Convirtiendo la columna Salida de object a date en cleaned_df_anfitrion

# Convertir la columna 'Salida' a formato datetime
cleaned_df_anfitrion['Salida'] = pd.to_datetime(cleaned_df_anfitrion['Salida'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_anfitrion['Salida'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Salida'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_anfitrion['Salida'].dtype)
```

```python
#################    CoHost    ###############
# Convirtiendo la columna Fecha de Reservación de object a date en cleaned_df_cohost

# Si en el formulario de Excel se deja vacio la casilla de Fecha de Reservación, esta se guarda como "00/01/1900" en la BBDD
# Convertir a string y limpiar espacios
cleaned_df_cohost["Fecha de Reservación"] = cleaned_df_cohost["Fecha de Reservación"].astype(str).str.strip()

# Reemplazar el valor "00/01/1900" por NaT antes de la conversión
cleaned_df_cohost["Fecha de Reservación"] = cleaned_df_cohost["Fecha de Reservación"].replace("00/01/1900", pd.NA)

# Convert 'Fecha de Reservación' to datetime format (YYYY-MM-DD)
cleaned_df_cohost['Fecha de Reservación'] = pd.to_datetime(
    cleaned_df_cohost['Fecha de Reservación'], 
    format='%d-%m-%Y', 
    errors='coerce'
)

# Verify if there are NaT values after conversion, many values will come as NaT values from the excel form.
if cleaned_df_cohost['Fecha de Reservación'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Fecha de Reservación'].isna()])

# Display the column type after conversion
print(cleaned_df_cohost['Fecha de Reservación'].dtype)





#################    Anfitrion    ###############
# Convirtiendo la columna Fecha de Reservación de object a date en cleaned_df_anfitrion

# Si en el formulario de Excel se deja vacio la casilla de Fecha de Reservación, esta se guarda como "00/01/1900" en la BBDD
# Convertir a string y limpiar espacios
cleaned_df_anfitrion["Fecha de Reservación"] = cleaned_df_anfitrion["Fecha de Reservación"].astype(str).str.strip()

# Reemplazar el valor "00/01/1900" por NaT antes de la conversión
cleaned_df_anfitrion["Fecha de Reservación"] = cleaned_df_anfitrion["Fecha de Reservación"].replace("00/01/1900", pd.NA)

# Convert 'Fecha de Reservación' to datetime format (YYYY-MM-DD)
cleaned_df_anfitrion['Fecha de Reservación'] = pd.to_datetime(
    cleaned_df_anfitrion['Fecha de Reservación'], 
    format='%d-%m-%Y', 
    errors='coerce'
)

# Verify if there are NaT values after conversion, many values will come as NaT values from the excel form.
if cleaned_df_anfitrion['Fecha de Reservación'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Fecha de Reservación'].isna()])

# Display the column type after conversion
print(cleaned_df_anfitrion['Fecha de Reservación'].dtype)



#Revisando todos los cambios de Date que realizamos
print(cleaned_df_cohost.head())
print(cleaned_df_anfitrion.head())
```
