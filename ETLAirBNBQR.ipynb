```python
# pip install pandas

import pandas as pd
import numpy as np

file_path = "BBDD Huespedes (Macros).xlsm"

# Leer las dos hojas específicas en un diccionario de DataFrames
sheets_dict = pd.read_excel(file_path, sheet_name=["BBDD CoHost", "BBDD Anfitrion"], engine="openpyxl")

# Extraer cada hoja en un DataFrame separado
df_cohost = sheets_dict["BBDD CoHost"]
df_anfitrion = sheets_dict["BBDD Anfitrion"]

# Mostrar las primeras filas de cada DataFrame para verificar
print("BBDD CoHost:")
print(df_cohost.head())
df_cohost.info()

print("\nBBDD Anfitrion:")
print(df_anfitrion.head())
df_anfitrion.info()
```

```python
#################    CoHost    ###############

# Convirtiendo la columna Ingreso de object a date Tanto para df_cohost
# Create a copy of cleaned_df_cohost to avoid modifying the original dataframe
cleaned_df_cohost = df_cohost.copy()

# Convertir la columna 'Ingreso' a formato datetime
cleaned_df_cohost['Ingreso'] = pd.to_datetime(cleaned_df_cohost['Ingreso'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_cohost['Ingreso'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Ingreso'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_cohost['Ingreso'].dtype)




#################    Anfitrion    ###############
# Convirtiendo la columna Ingreso de object a date Tanto para df_anfitrion
# Create a copy of cleaned_df_anfitrion to avoid modifying the original dataframe
cleaned_df_anfitrion = df_anfitrion.copy()

# Convertir la columna 'Ingreso' a formato datetime
cleaned_df_anfitrion['Ingreso'] = pd.to_datetime(cleaned_df_anfitrion['Ingreso'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_anfitrion['Ingreso'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Ingreso'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_anfitrion['Ingreso'].dtype)
```

```python
#################    CoHost    ###############
# Convirtiendo la columna Salida de object a date en cleaned_df_cohost

# Convertir la columna 'Salida' a formato datetime
cleaned_df_cohost['Salida'] = pd.to_datetime(cleaned_df_cohost['Salida'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_cohost['Salida'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Salida'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_cohost['Salida'].dtype)



#################    Anfitrion    ###############
# Convirtiendo la columna Salida de object a date en cleaned_df_anfitrion

# Convertir la columna 'Salida' a formato datetime
cleaned_df_anfitrion['Salida'] = pd.to_datetime(cleaned_df_anfitrion['Salida'], format='%d/%m/%Y', errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_anfitrion['Salida'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Salida'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_anfitrion['Salida'].dtype)
```

```python
#################    CoHost    ###############
# Convirtiendo la columna Fecha de Reservación de object a date en cleaned_df_cohost

# Si en el formulario de Excel se deja vacio la casilla de Fecha de Reservación, esta se guarda como "00/01/1900" en la BBDD
# Convertir a string y limpiar espacios
cleaned_df_cohost["Fecha de Reservación"] = cleaned_df_cohost["Fecha de Reservación"].astype(str).str.strip()

# Reemplazar el valor "00/01/1900" por NaT antes de la conversión
cleaned_df_cohost["Fecha de Reservación"] = cleaned_df_cohost["Fecha de Reservación"].replace("00/01/1900", pd.NA)

# Convert 'Fecha de Reservación' to datetime format (YYYY-MM-DD)
cleaned_df_cohost['Fecha de Reservación'] = pd.to_datetime(
    cleaned_df_cohost['Fecha de Reservación'], 
    format='%d-%m-%Y', 
    errors='coerce'
)

# Verify if there are NaT values after conversion, many values will come as NaT values from the excel form.
if cleaned_df_cohost['Fecha de Reservación'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_cohost[cleaned_df_cohost['Fecha de Reservación'].isna()])

# Display the column type after conversion
print(cleaned_df_cohost['Fecha de Reservación'].dtype)





#################    Anfitrion    ###############
# Convirtiendo la columna Fecha de Reservación de object a date en cleaned_df_anfitrion

# Si en el formulario de Excel se deja vacio la casilla de Fecha de Reservación, esta se guarda como "00/01/1900" en la BBDD
# Convertir a string y limpiar espacios
cleaned_df_anfitrion["Fecha de Reservación"] = cleaned_df_anfitrion["Fecha de Reservación"].astype(str).str.strip()

# Reemplazar el valor "00/01/1900" por NaT antes de la conversión
cleaned_df_anfitrion["Fecha de Reservación"] = cleaned_df_anfitrion["Fecha de Reservación"].replace("00/01/1900", pd.NA)

# Convert 'Fecha de Reservación' to datetime format (YYYY-MM-DD)
cleaned_df_anfitrion['Fecha de Reservación'] = pd.to_datetime(
    cleaned_df_anfitrion['Fecha de Reservación'], 
    format='%d-%m-%Y', 
    errors='coerce'
)

# Verify if there are NaT values after conversion, many values will come as NaT values from the excel form.
if cleaned_df_anfitrion['Fecha de Reservación'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_anfitrion[cleaned_df_anfitrion['Fecha de Reservación'].isna()])

# Display the column type after conversion
print(cleaned_df_anfitrion['Fecha de Reservación'].dtype)



#Revisando todos los cambios de Date que realizamos
print(cleaned_df_cohost.head())
print(cleaned_df_anfitrion.head())
```

```python
# Strip whitespace and ensure string type
cleaned_df_cohost['Contacto'] = cleaned_df_cohost['Contacto'].astype(str).str.strip()

# Contacto 0 values should be Nan.
cleaned_df_cohost['Contacto'] = cleaned_df_cohost['Contacto'].replace("0", np.nan)

# Finding out if there are any missing values.
print(cleaned_df_cohost.isna().sum())

# We have 24 missing values in Contacto, this values should be null values.
print(cleaned_df_cohost['Contacto'].value_counts()) 

# The DataFrame lenght is 26
print(len(cleaned_df_cohost))
```

```python
# Adding Property Type to further our analysis, to underestand what type of properties perform better, mapping the values from 'Propiedad' 

#################    CoHost    ###############
# Define the mapping dictionary
property_type_mapping = {
    'MACORORO 12 - 12F': '1 Dormitorio',
    'YOU - 316': 'Monoambiente',
    'SKY ELITE - 320': 'Monoambiente',
    'SKY ELITE - 132': '1 Dormitorio',
    'SKY ELITE - 229': 'Monoambiente',
    'SKY ELITE - 115': 'Monoambiente',
    'TORRES SIRARI - 5A': '1 Dormitorio',
    'MACORORO 8': '1 Dormitorio',
    'MACORORO 15 - 8G': '3 Dormitorios',
    'SKY ELITE - 524': 'Monoambiente',
    'SKY ELITE - 323': 'Monoambiente',
    'HOTEL RADISSON - SENIOR SUITE': 'Senior Suite',
    'SKY ELITE - 510': 'Monoambiente',
    'SKY LUXIA - 513': 'Monoambiente'
}

# Create the new column 'Property Type' using the mapping dictionary
cleaned_df_cohost['Property Type'] = cleaned_df_cohost['Propiedad'].map(property_type_mapping)

cleaned_df_anfitrion['Property Type'] = cleaned_df_anfitrion['Propiedad'].map(property_type_mapping)

# Display the updated DataFrame
print(cleaned_df_cohost[['Propiedad', 'Property Type']].head())
print(cleaned_df_anfitrion[['Propiedad', 'Property Type']].head())
```

```python
# Adding if the properties are of our own (family) or one of our clients that we manage them for

#################    CoHost    ###############
# Define the mapping dictionary
property_management_mapping = {
    'MACORORO 12 - 12F': 'Propia',
    'YOU - 316': 'Administrada',
    'SKY ELITE - 320': 'Administrada',
    'SKY ELITE - 132': 'Administrada',
    'SKY ELITE - 229': 'Propia',
    'SKY ELITE - 115': 'Administrada',
    'TORRES SIRARI - 5A': 'Administrada',
    'MACORORO 8': 'Propia',
    'MACORORO 15 - 8G': 'Propia',
    'SKY ELITE - 524': 'Administrada',
    'SKY ELITE - 323': 'Administrada',
    'HOTEL RADISSON - SENIOR SUITE': 'Administrada',
    'SKY ELITE - 510': 'Administrada',
    'SKY LUXIA - 513': 'Administrada'
}

# Create the new column 'Gestion' using the mapping dictionary
cleaned_df_cohost['Management'] = cleaned_df_cohost['Propiedad'].map(property_management_mapping)

cleaned_df_anfitrion['Management'] = cleaned_df_anfitrion['Propiedad'].map(property_management_mapping)

# Display the updated DataFrame
print(cleaned_df_cohost[['Propiedad', 'Management']].head())
print(cleaned_df_anfitrion[['Propiedad', 'Management']].head())
```

```python
# Asigning the correct values to the 'Propietario' column, not needed for the cleaned_df_anfitrion
# Define a mapping to replace values in 'Propietario'
propietario_name_mapping = {
    'Dr._Ortiz': 'Dr. Ortiz',
    'Boris_Ayviri': 'Boris Ayviri',
    'Lic._Freddy_Soria': 'Lic. Freddy Soria',
    'Roberto_Vargas': 'Roberto Vargas',
    'Gabriela_Salama': 'Gabriela Salama',
    'Betty_Vargas': 'Betty Vargas'
}

# Apply the replacement on the 'Propietario' column
cleaned_df_cohost['Propietario'] = cleaned_df_cohost['Propietario'].replace(propietario_name_mapping)

# cleaned_df_anfitrion['Propietario'] = cleaned_df_anfitrion['Propietario'].replace(propietario_name_mapping)

# verify the result
print(cleaned_df_cohost['Propietario'].unique())
print(cleaned_df_anfitrion['Propietario'].unique())
```

```python
# Crear copia para no modificar los originales
df_cohost_combined = cleaned_df_cohost.copy()
df_anfitrion_combined = cleaned_df_anfitrion.copy()

# Calcular 'Ingresos Totales BOB' como suma de 'Comision BOB' + 'Limpieza BOB'
df_cohost_combined['Ingresos Totales BOB'] = df_cohost_combined['Comision BOB'] + df_cohost_combined['Limpieza BOB']

# Definir columnas comunes para mantener
columns_to_keep = [
    'Nombre del huésped', 'Contacto', 'Número de adultos', 'Número de niños',
    'Ingreso', 'Salida', 'Número de noches', 'Fecha de Reservación',
    'Propietario', 'Propiedad', 'Ingresos Totales BOB', 'Limpieza BOB',
    'Property Type', 'Management'
]

# Seleccionar columnas comunes de ambos dataframes
df_cohost_common = df_cohost_combined[columns_to_keep]
df_anfitrion_common = df_anfitrion_combined[columns_to_keep]

# Concatenar los dos DataFrames
df_combined_BOB = pd.concat([df_cohost_common, df_anfitrion_common], ignore_index=True)

# Verificar la estructura
print("Columnas del DataFrame combinado:", df_combined_BOB.columns.tolist())
print("Total de filas combinadas:", len(df_combined_BOB))
print(df_combined_BOB.head())
```
