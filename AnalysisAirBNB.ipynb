```python
import pandas as pd

df_Combined_BOB = pd.read_csv('cleaned_BOB_df.csv')
cleaned_Outliers_df = pd.read_csv('cleaned_Outliers.csv')

print(df_Combined_BOB.info())
print(cleaned_Outliers_df.info())
```

```python
# Crear Código de confirmación artificial en df_combined_BOB
df_combined_BOB = df_Combined_BOB.copy()
df_combined_BOB["Código de confirmación"] = ["COH{:05d}".format(i) for i in range(1, len(df_combined_BOB) + 1)]

# Definir columnas comunes deseadas (sin columnas de dinero)
columns_to_keep = [
    "Código de confirmación", "Nombre del huésped", "Contacto", "Número de adultos",
    "Número de niños", "Ingreso", "Salida", "Número de noches", "Fecha de Reservación",
    "Propiedad", "Property Type", "Management", "Propietario"
]

# Seleccionar columnas deseadas en ambos DFs
df_BOB_clean = df_combined_BOB[columns_to_keep]
df_USD_clean = cleaned_Outliers_df[columns_to_keep]

# 4. Concatenar
df_analysis = pd.concat([df_BOB_clean, df_USD_clean], ignore_index=True)

print(df_analysis.info())
print(df_analysis[['Ingreso', 'Salida', 'Fecha de Reservación']])
```

```python
# Convirtiendo la columna Ingreso de object a date
# Create a copy of df_analysis to avoid modifying the original dataframe
cleaned_df_analysis = df_analysis.copy()

# Limpiar espacios en blanco y forzar string
cleaned_df_analysis['Ingreso'] = cleaned_df_analysis['Ingreso'].astype(str).str.strip()

cleaned_df_analysis['Ingreso'] = pd.to_datetime(cleaned_df_analysis['Ingreso'], errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_analysis['Ingreso'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_analysis[cleaned_df_analysis['Ingreso'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_analysis['Ingreso'].dtype)
print(cleaned_df_analysis.info())
```

```python
# Convirtiendo la columna Salida de object a date en cleaned_df_analysis

# Limpiar espacios en blanco y forzar string
cleaned_df_analysis['Salida'] = cleaned_df_analysis['Salida'].astype(str).str.strip()

cleaned_df_analysis['Salida'] = pd.to_datetime(cleaned_df_analysis['Salida'], errors='coerce')

# Verificar si hay valores nulos después de la conversión
if cleaned_df_analysis['Salida'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_analysis[cleaned_df_analysis['Salida'].isna()])

# Mostrar el tipo de la columna después de la conversión
print(cleaned_df_analysis['Salida'].dtype)
print(cleaned_df_analysis.info())
```

```python
# Convirtiendo la columna Fecha de Reservación de object a date en cleaned_df_analysis

# Si en el formulario de Excel se deja vacio la casilla de Fecha de Reservación, esta se guarda como "00/01/1900" en la BBDD
# Convertir a string y limpiar espacios
cleaned_df_analysis["Fecha de Reservación"] = cleaned_df_analysis["Fecha de Reservación"].astype(str).str.strip()

cleaned_df_analysis['Fecha de Reservación'] = pd.to_datetime(cleaned_df_analysis['Fecha de Reservación'], errors='coerce')

# Verify if there are NaT values after conversion, many values will come as NaT values from the excel form.
if cleaned_df_analysis['Fecha de Reservación'].isna().any():
    print("Existen valores no convertidos a datetime. Revisa los siguientes datos:")
    print(cleaned_df_analysis[cleaned_df_analysis['Fecha de Reservación'].isna()])

# Display the column type after conversion
print(cleaned_df_analysis['Fecha de Reservación'].dtype)
print(cleaned_df_analysis.info())
```

```python
# For global Analysis
print(cleaned_df_analysis.info())

# Night's book DataFrame, to see the Ocupancy rate per property and the average # of Nights per property. We'll also see average
# Booking Lead Time per property
ocupancy_property_df = cleaned_df_analysis[['Ingreso', 'Salida', 'Número de noches', 'Fecha de Reservación', 'Propiedad', 'Property Type']].copy()

# Agregar Mes y Año
ocupancy_property_df["Month"] = ocupancy_property_df["Ingreso"].dt.month
ocupancy_property_df["Year"] = ocupancy_property_df["Ingreso"].dt.year

# Calculate Booking Lead Time (Days Between Booking and Check-In)
ocupancy_property_df["Lead Time"] = (ocupancy_property_df["Ingreso"] - ocupancy_property_df["Fecha de Reservación"]).dt.days
#print(ocupancy_property_df.head())

# ------------------ OCCUPANCY RATE AND BOOKING LEAD TIME FOR PROPERTY ------------------
# Aggregate Nights Booked per Property, Month, Year
occupancy_prop_df = ocupancy_property_df.groupby(["Propiedad", "Month", "Year"]).agg({
    "Número de noches": "sum",
    "Lead Time": "mean"}).reset_index()
occupancy_prop_df.rename(columns={"Número de noches": "Total Nights Booked per Property", "Lead Time": "Avg Lead Time per Property"}, inplace=True)

# Calculate Total Available Nights Based on Month
import calendar
occupancy_prop_df["Total Available Nights"] = occupancy_prop_df["Month"].apply(lambda x: calendar.monthrange(2024, x)[1])

# Calculate Occupancy Rate for Each Property
occupancy_prop_df["Occupancy Rate (%)"] = (occupancy_prop_df["Total Nights Booked per Property"] / occupancy_prop_df["Total Available Nights"]) * 100
print(occupancy_prop_df.head())

# ------------------ OCCUPANCY RATE FOR PROPERTY TYPE ------------------
occupancy_type_df = ocupancy_property_df.groupby(["Property Type", "Month", "Year"]).agg({
    "Número de noches": "sum",
    "Lead Time": "mean"}).reset_index()
occupancy_type_df.rename(columns={"Número de noches": "Total Nights Booked per Property Type", "Lead Time": "Avg Lead Time per Property Type"}, inplace=True)

# Calculate Total Available Nights Based on Month
occupancy_type_df["Total Available Nights"] = occupancy_type_df["Month"].apply(lambda x: calendar.monthrange(2024, x)[1])

# Calculate Occupancy Rate for Each Property Type
occupancy_type_df["Occupancy Rate (%)"] = (occupancy_type_df["Total Nights Booked per Property Type"] / occupancy_type_df["Total Available Nights"]) * 100
print(occupancy_type_df.head())


# One will be used for day to day transactions, such as occupancy rate. Used for: Occupancy rates, Booking lead time, Lenght Stay trends, Loyalty trends
# The other will be used for aggregated functions for property comparisons, between months. Used for: Earnings analysis, Demographic Analysis, Seasonal Analysis
```

```python
# There was an interpretation mistake done while making the occupancy rate analysis.
# There are months were there were no bookings, are this months are actually not shown in the current dataframe.
# All months must appear and if no bookings are registered, there should be a 0% occupancy rate
# Months should start count from the first month the first property was offered, in the "Ingreso" column

from datetime import datetime

# Get current year and month (real-time reference)
current_year = datetime.today().year
current_month = datetime.today().month

# Get the first 'Ingreso' (first booking date) for each property
first_offer_date = ocupancy_property_df.groupby("Propiedad")["Ingreso"].min().reset_index()
first_offer_date.rename(columns={"Ingreso": "First Ingreso"}, inplace=True)

first_offer_date["First Year"] = first_offer_date["First Ingreso"].dt.year
first_offer_date["First Month"] = first_offer_date["First Ingreso"].dt.month

# Get the full date range from the first to the last 'Ingreso'
min_year, min_month = ocupancy_property_df["Ingreso"].min().year, ocupancy_property_df["Ingreso"].min().month

from itertools import product

# Create a full list of month-year combinations
all_months_years = pd.DataFrame(
    list(product(range(min_year, current_year + 1), range(1, 13))),
    columns=["Year", "Month"]
)

# Filter to ensure we don't include months beyond the current real-time month
all_months_years = all_months_years[
    (all_months_years["Year"] < current_year) | 
    ((all_months_years["Year"] == current_year) & (all_months_years["Month"] <= current_month))
]

# ------------------ COMPLETE OCCUPANCY DATA PER PROPERTY ------------------

# Get all unique properties
all_properties = ocupancy_property_df[['Propiedad']].drop_duplicates()

# Cross join properties with all months/years
full_prop_df = all_properties.merge(all_months_years, how="cross")

# Merge to get the first offer date per property
full_prop_df = full_prop_df.merge(first_offer_date, on="Propiedad", how="left")

# Filter out months before the property was offered
full_prop_df = full_prop_df[
    (full_prop_df["Year"] > full_prop_df["First Year"]) | 
    ((full_prop_df["Year"] == full_prop_df["First Year"]) & (full_prop_df["Month"] >= full_prop_df["First Month"]))
]

# Drop unnecessary columns
full_prop_df2 = full_prop_df.drop(columns=["First Ingreso", "First Year", "First Month"]).copy()

# Merge with existing occupancy data into a new DataFrame
occupancy_prop_df2 = full_prop_df2.merge(occupancy_prop_df, on=["Propiedad", "Year", "Month"], how="left")

# Fill missing values with 0 for occupancy-related metrics
occupancy_prop_df2["Total Nights Booked per Property"] = occupancy_prop_df2["Total Nights Booked per Property"].fillna(0)
occupancy_prop_df2["Total Available Nights"] = occupancy_prop_df2["Month"].apply(lambda x: calendar.monthrange(2024, x)[1])
occupancy_prop_df2["Occupancy Rate (%)"] = (occupancy_prop_df2["Total Nights Booked per Property"] / occupancy_prop_df2["Total Available Nights"]) * 100

print(occupancy_prop_df2.head())
```

```python
# We continue with the occupancy_type_df, where apart from the mistake made with occupancy_prop_df, there were also months
# were the occupancy rate was more than 100%, this is due because there might be more than 1 porperty as a property type,
# meaning that when we sum up the booked nights, we might get more than the total aviable nights. Total aviable nights should be
# dynamic, depending on the number of properties in a property type and month


# ------------------ COMPLETE OCCUPANCY DATA PER PROPERTY TYPE ------------------

# Map each property to its property type
property_type_mapping = ocupancy_property_df[['Propiedad', 'Property Type']].drop_duplicates()

# Merge to get the first booking date per property type
property_first_booking = first_offer_date.merge(property_type_mapping, on="Propiedad", how="left")

# Get first offer date per property type (Earliest Date for each Property Type)
property_first_booking = property_first_booking.groupby("Property Type")["First Ingreso"].min().reset_index()
property_first_booking["First Year"] = property_first_booking["First Ingreso"].dt.year
property_first_booking["First Month"] = property_first_booking["First Ingreso"].dt.month

# Get all unique property types
all_property_types = ocupancy_property_df[['Property Type']].drop_duplicates()

# Cross join property types with all months/years (Reuse 'all_months_years')
full_type_df = all_property_types.merge(all_months_years, how="cross")

# Merge with the first available booking date per property type
full_type_df = full_type_df.merge(property_first_booking, on="Property Type", how="left")


# FILTER OUT MONTHS BEFORE THE FIRST TIME A PROPERTY TYPE WAS OFFERED
full_type_df = full_type_df[
    (full_type_df["Year"] > full_type_df["First Year"]) | 
    ((full_type_df["Year"] == full_type_df["First Year"]) & (full_type_df["Month"] >= full_type_df["First Month"]))
]

# Drop unnecessary columns
full_type_df = full_type_df.drop(columns=["First Ingreso", "First Year", "First Month"]).copy()

# Count available properties per type each month (Reuse 'ocupancy_property_df')
available_properties_per_month = ocupancy_property_df.groupby(["Property Type", "Year", "Month"])["Propiedad"].nunique().reset_index()
available_properties_per_month.rename(columns={"Propiedad": "Available Properties"}, inplace=True)

# Merge to get the number of properties available each month per type
full_type_df = full_type_df.merge(available_properties_per_month, on=["Property Type", "Year", "Month"], how="left")

# Fill NaN values (months with 0 properties should be counted as 0)
full_type_df["Available Properties"] = full_type_df["Available Properties"].fillna(0)

# Merge with existing occupancy data (Reuse 'occupancy_type_df')
occupancy_type_df2 = full_type_df.merge(occupancy_type_df, on=["Property Type", "Year", "Month"], how="left")

# Fill missing values with 0 for total nights booked
occupancy_type_df2["Total Nights Booked per Property Type"] = occupancy_type_df2["Total Nights Booked per Property Type"].fillna(0)

# Ensure available nights are calculated correctly even if there are no bookings
occupancy_type_df2["Total Available Nights"] = occupancy_type_df2.apply(
    lambda row: max(row["Available Properties"], 1) * calendar.monthrange(row["Year"], row["Month"])[1], axis=1
)

# Compute Corrected Occupancy Rate
occupancy_type_df2["Occupancy Rate (%)"] = (
    occupancy_type_df2["Total Nights Booked per Property Type"] / occupancy_type_df2["Total Available Nights"]
) * 100

# Ensure no occupancy rate exceeds 100%
occupancy_type_df2["Occupancy Rate (%)"] = occupancy_type_df2["Occupancy Rate (%)"].clip(upper=100)

print(occupancy_type_df2.head())
```

```python
# Customer Analysis
# I'll be creating a customer_loyalty_df to see which customers re used our services, percent of customers that only used once our 
# services, and percent of customers that used more than once our services. Demography of our clients

customer_df = cleaned_df_analysis[['Código de confirmación', 'Nombre del huésped', 'Contacto', 'Número de adultos', 'Número de niños', 'Propiedad', 'Property Type']].copy()

# Crear un identificador único por huésped (puede ser nombre o contacto si falta el nombre)
customer_df["Guest Identifier"] = customer_df.apply(lambda x: x["Nombre del huésped"] if pd.notna(x["Nombre del huésped"]) else x["Contacto"], axis=1)

# Agrupar por huésped y contar la cantidad de reservas (Total Bookings)
customer_loyalty_df = customer_df.groupby("Guest Identifier").agg({"Código de confirmación": "nunique"}).reset_index()

# Renombrar columnas
customer_loyalty_df.rename(columns={"Código de confirmación": "Total Bookings"}, inplace=True)
customer_loyalty_df = customer_loyalty_df.sort_values(by="Total Bookings", ascending=False)
print(customer_loyalty_df.head())
print(customer_loyalty_df.info())
print(customer_loyalty_df.describe())
```

```python
# Customer Demographic

# ------------------ CUSTOMER DEMOGRAPHICS PER PROPERTY ------------------
customer_demograph_property_df = customer_df.groupby(["Propiedad"]).agg({
    "Número de adultos": "sum",
    "Número de niños": "sum",
    "Código de confirmación": "count"  # Total bookings per property
}).reset_index()

customer_demograph_property_df.rename(columns={"Confirmation Code": "Total Bookings"}, inplace=True)
print(customer_demograph_property_df)

# ------------------ CUSTOMER DEMOGRAPHICS PER PROPERTY TYPE ------------------
customer_demograph_type_df = customer_df.groupby(["Property Type"]).agg({
    "Número de adultos": "sum",
    "Número de niños": "sum",
    "Código de confirmación": "count"
}).reset_index()

customer_demograph_type_df.rename(columns={"Confirmation Code": "Total Bookings"}, inplace=True)
print(customer_demograph_type_df)
```
